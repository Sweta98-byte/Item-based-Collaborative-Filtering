```python
def convert_matrix(data):
    user = data['userId'].unique()
    item = data['itemId'].unique()
    mat = pd.DataFrame(0,index =user,columns = item)
    leng = data['userId'].count()
    for i in range(leng):
        in1 = data['userId'][i]
        in2 = data['itemId'][i]
        mat[in2][in1] = data['rating'][i]
    return mat
    
```

```python
mat = convert_matrix(data)
```

```python
#Train and test Splitting
```

```python
def train_test(mat):
    test = pd.DataFrame(0.0,columns = mat.columns ,index = mat.index)
    user = mat.index
    for i in user:
        l = mat.iloc[i-1,:]
        l = l[l!=0]
        t = l.sample(frac =0.2)
        for j in t.index:
            rat = mat[j][i]
            test[j][i] = rat
            mat[j][i] =0
    return test,mat
```

```python
data_test ,data_train = train_test(mat)
```

#Calculating Similarity

```python
from math import sqrt
def adj_cosine(data_train):
    item = data_train.columns
    user = data_train.index
    mat = pd.DataFrame(0.00,index = item,columns=item)
    for i in item:
        mat[i][i] =1
    for k in range(item.size):
        a = item[k]
        for i in range(k+1,item.size):
            b = item[i]
            sum =0
            mul1 =0
            mul2 =0
            c=0
            set_c=((data_train.iloc[:,k] !=0)*(data_train.iloc[:,i]!=0))
            df = set_c[set_c == True]
            for j in df.index:
                m = data_train['mean'][j]
                rat1 = data_train[a][j]
                rat2 = data_train[b][j]
                sum = sum+(rat1-m)*(rat2-m)
                mul1 = mul1+(rat1-m)**2
                mul2 = mul2+(rat2-m)**2
            if mul1!=0 and mul2!=0:   
                c = sum/sqrt(mul1*mul2)
            mat[a][b] = c
            mat[b][a] = mat[a][b]
        print(k)
    return mat
```

```python
sim = adj_cosine(data_train)
```

```python
def prediction(test,sim,train):
    user = test.index
    item = test.columns
    mat = pd.DataFrame(0.00,index = user,columns = item)
    sum =0
    weight =0
    f =0;
    for i in user:
        for j in item:
            a = test[j][i]
            if a==0:
                mat[j][i] =0
            else:
#                 print(test[j][i])tes
                sum =0
                weight =0
                for k in item:
                    b = train[k][i]
                    if k==j or sim[j][k]<0 :
                        continue
                    if b!=0:
                        sum = sum + sim[j][k]*b
                        weight = weight+abs(sim[j][k])
#                 print(sum,weight)
                if weight==0:
                    mat[j][i] =0
                else:
                    mat[j][i] = sum/weight
#                     print(mat[j][i])
        f+=1
        print(f)
    return mat
```

```python
def evaluation(actual,pred,k):
    l = len(actual)
    tp =0
    fp =0
    tn =0
    fn =0
    for i in range(l):
        if actual[i]>=k and pred[i]>=k:
            tp+=1
        elif actual[i]<k and pred[i]<k:
            tn+=1
        elif actual[i]<k and pred[i]>=k:
            fp+=1
        elif actual[i]>=k and pred[i]<k:
            fn+=1
    prec = tp/(tp+fp)
    reca = tp/(tp+fn)
    return prec,reca
    
```

```python
def create_class(actual,pred,k):
    y1 =[]
    yp1 =[]
    l = len(actual)
    for i in range(l):
        if actual[i]>=k:
            y1.append(1)
        else:
            y1.append(0)
        if pred[i]>=k:
            yp1.append(1)
        else:
            yp1.append(0)
    return y1,yp1
    
```

```python
from sklearn import metrics
y1,yp1 = create_class(y,yp,4)
# len(yp1)
print(metrics.confusion_matrix(y1,yp1))
```
